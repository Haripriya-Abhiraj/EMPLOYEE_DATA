## How many are there in each Team and the percentage splitting with respect to the total employees.

```python
Total_Team_Count=df["Team"].value_counts()
print("Total count :")
print(Total_Team_Count)
```

```python
Percentage =Total_Team_Count/len(df)*100
print("Team_Percentage:")
print(Percentage)
```

## Find from which age group most of the employees belong to


```python
df['Age']
```

```python
AGE_COUNT=df['Age'].value_counts()
print(AGE_COUNT)
```

```python
most_age_count1=AGE_COUNT.idxmin()
most_age_count=AGE_COUNT.idxmax()
print(most_age_count,",",most_age_count1)
```

```python
from matplotlib import pyplot as plt
bins = np.arange(18,40)

plt.figure(figsize=(8,5))

plt.hist(df.Age, bins=bins, color='green')

plt.xticks(bins)

plt.ylabel('Number Of Employees')
plt.xlabel('Age')
plt.title('DAGE GROUP OF DISTRIBUTION OF EMPLOYEES')



plt.show()
```

## Find out under which team and position, spending in terms of salary is high.

```python
highest_salary=df.groupby(["Team","Position"])["Salary"].idxmax().sum()
print(highest_salary)
```

## Find if there is any correlation between age and salary , represent it visually.

```python

plt.scatter(df["Age"],df["Salary"])
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("CORRELATION BETWEEN AGE AND SALARY")
plt.show()
```
